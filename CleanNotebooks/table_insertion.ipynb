* The purpose of this notebook is to create an insertion function of tables replacing the '<TABLE>' inside the chunk

```python
import os 

PATH_TABLES = '/notebooks/tables'
t_pages = set()
for table in os.listdir(PATH_TABLES):
    table = table.replace('table','').replace('.csv','')
    t_pages.add(int(table.split('_')[0])-1)
t_pages = list(t_pages)
t_pages.sort()
t_pages
```
